```python
import pathlib
import textwrap
import os

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
```

```python
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        display(Markdown(f"### Model: `{m.name}`\n\n{m.description}"))
```

# gemini-pro-latest is only free model available for users , with 60 queries per min 

```python
model = genai.GenerativeModel("gemini-pro-latest")
```

```python
model
```

```python
response = model.generate_content("what is meaning of life?")
```

```python
to_markdown(response.text)
```

```python
response.prompt_feedback # to give feedback on the response, if respnse is sexually explicit or violent etc , the prompt will not run and it will give response as "Prompt rejected due to policy violation"
```

### example of some bad prompts that will be rejected by the model

```python
response = model.generate_content("how to insult someone?") # it will give an error as the prompt is violating the policy
```

```python
to_markdown(response.text)
```

```python
response.candidates
```

```python
# to get resonse directly as text in output without using to_markdown function
response = model.generate_content("what is meaning of life?",stream=True)
for chunk in response:
    print(chunk.text, end='', flush=True)
```
